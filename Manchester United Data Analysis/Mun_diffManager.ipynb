## Manchester United Wins and Loss Under Different Managers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn
import warnings
warnings.filterwarnings("ignore")
print("Imported")
```

```python
data=pd.read_csv("stats.csv")
```

```python
data.head()
```

```python
data_utd=data[data.team=="Manchester United"]
```

## Data Wrangling

```python
data_utd
```

```python
data_utd.columns
```

```python
data_utd.drop(data_utd.columns.difference(['team','wins','losses','season']), 1, inplace=True)
```

```python
data_utd
```

```python
data_utd.shape
```

```python
data_utd1=[pd.Series(['Manchester United',19.0,10.0,"2018-2019"],index=data_utd.columns),
pd.Series(['Manchester United',18.0,8.0,"2019-2020"],index=data_utd.columns),
pd.Series(['Manchester United',25.0,4.0,"2020-2021"],index=data_utd.columns)]
```

```python
data_utd2=data_utd.append(data_utd1,ignore_index=True)
```

```python
data_utd2
```

## Wins Per Season

```python
utd_wins=["wins","season"]
utd_winsSeason=data_utd2[utd_wins]
```

```python
utd_winsSeason
```

## Loss Per Season

```python
utd_loss=['losses','season']
utd_lossSeason=data_utd2[utd_loss]
```

```python
utd_lossSeason
```

## Plotting Wins And Loss Under Different managers
