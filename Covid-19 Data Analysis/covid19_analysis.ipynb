```python
covid_df=covid_df.rename(columns={'State/UnionTerritory':'States'},inplace=False)
```

## Confirmed Cases :

```python
plt.figure(figsize=(10,5))
plt.title("Confirmed Cases")
plt.plot(covid_df['Confirmed'])
plt.show()
```

## Death Cases :

```python
plt.figure(figsize=(10,5))
plt.title("Death Cases")
plt.plot(covid_df["Deaths"])
plt.show()
```

## Cured Cases:

```python
plt.figure(figsize=(10,5))
plt.title("Cases Cured")
plt.plot(covid_df["Cured"])
plt.show()
```

## Cleaning the States DataFrame:

```python
covid_df['States'].unique()
```

```python
covid_df=covid_df.replace('Telangana','Telangana')
covid_df=covid_df.replace('Telengana***','Telangana')
covid_df=covid_df.replace('Telangana***','Telangana')
covid_df=covid_df.replace('Maharashtra***','Maharashtra')
covid_df=covid_df.replace('Punjab***','Punjab')
covid_df=covid_df.replace('Chandigarh***','Chandigarh')

```

```python
covid_df['States'].unique()
```

```python
drop_indexvalue=covid_df[covid_df['States']=='Cases being reassigned to states'].index
covid_df.drop(drop_indexvalue,inplace=True)
```

```python
covid_df['States'].unique()
```

```python
drop_indexvalue1=covid_df[covid_df['States']=='Unassigned'].index
covid_df.drop(drop_indexvalue1,inplace=True)
```

```python
covid_df['States'].unique()
```

## Data Count StateWise:

```python
plt.figure(figsize=(10,20))
plt.title("Data Count Per State")

sns.countplot(data=covid_df,y='States',order=covid_df['States'].value_counts().index,color='yellow')
plt.show()
```

```python
covid_df.describe()
```

## Active Cases:

```python
covid_df['Active']=covid_df['Confirmed']-covid_df['Cured']-covid_df['Deaths']
covid_df['Active']
```

```python
x=covid_df['Confirmed'].sum()
print("Total Confirmed Cases are ",x)
y=covid_df['Deaths'].sum()
print("Total Death Cases are ",y)
z=covid_df['Cured'].sum()
print("Total Cured Cases are ",z)
active=x-y-z
print("Total Active Cases are ",active)
```

## Plotting Confirmed ,Death,Cured ,Active Case:

```python
plt.figure(figsize=(10,5))
plt.title("Confirmed ,Deaths,Cured,Active Cases in India")
plt.ylabel("Peoples(in crore)")
sns.barplot(x=['Confirmed','Death','Cured','active'],y=[x,y,z,active])
```

## Top Most Affected State:

```python
covid_df1=covid_df.tail(35)
covid_df2=covid_df1.sort_values(by='Confirmed',ascending=False).head(10)
```

```python
covid_df2
```

```python
plt.figure(figsize=(20,10))
plt.title("Most Affected States value ")
sns.barplot(x=covid_df2['States'],y=covid_df2['Confirmed'],data=covid_df2,color="#00CED1")
plt.ylabel("Confirmed(in millions)")
plt.show()
```

## Fatality-Ratio:

```python
covid_df2['Fatality-Ratio']=covid_df2['Deaths']/covid_df2['Confirmed']
```

```python
covid_df2['Fatality-Ratio']
```

```python
plt.figure(figsize=(20,10))
plt.title("10 Most Affected States Fatality-Ration Rate ")
sns.pointplot(x="States",y="Fatality-Ratio",data=covid_df2,color="#6A5ACD")
plt.show()
```

## Cured Ratio:

```python
covid_df2["Cured-Ratio"]=covid_df2["Cured"]/covid_df2["Confirmed"]
covid_df2["Cured-Ratio"]
```

```python
plt.figure(figsize=(20,10))
plt.title("States Cured Ratio")
sns.pointplot(x=covid_df2['States'],y=covid_df2["Cured-Ratio"],data=covid_df2,color="#D2691E")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.title("StateWise Active Cases")
sns.barplot(x=covid_df2["States"],y=covid_df2["Active"],data=covid_df2,color="#6495ED")
plt.show()
```
